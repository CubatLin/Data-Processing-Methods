```python
%reset -f
import pandas as pd
import numpy as np
from pandas.core.frame import DataFrame
import math
import scipy
import statistics as stat
import matplotlib.pyplot as plt
```

```python
#拿到一串陣列
test = np.random.choice(a=[0,1],size=20,p=[0.7,0.3])

#找出類別標記
test_1 = np.where(test==1)
test_1
```

3,4,6,7,9,附近有1的標記...

```python
test_kde = list(range(test.shape[0]))
```

```python
#O(N)
for i in test_1:
    test_kde.extend(i)
#O(N)
test_kde.sort()
```

```python
#Kernel Show Time
import scipy.stats as st
plt.rc('figure', figsize=(11.69,8.27))
kde = st.gaussian_kde(test_kde)
plt.plot(test_kde, kde.pdf(test_kde), label="PDF")
plt.legend(loc="upper left")
plt.ylabel('Probability')
plt.xlabel('Data')
plt.title("Histogram");
```
