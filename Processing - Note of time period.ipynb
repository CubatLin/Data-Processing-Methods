## Q1: Conduct DateTime

```python
data['event_time'] = pd.to_datetime(data['event_time'], format='%Y-%m-%d %H:%M:%S')
data.head(10)
```

```python
data = data.sort_values(['shopid','event_time','userid'])
data
```

```python
data['previous_time'] = data.groupby(['shopid'])['event_time'].shift(1)
```

```python
data.head(20)
```

```python
data['time_diff'] = data['event_time'] - data['previous_time']
data.head(20)
```

```python
data['time_diff_d'] = data['time_diff'].astype('timedelta64[D]')
data['time_diff_m'] = data['time_diff'].astype('timedelta64[m]')
data['time_diff_s'] = data['time_diff'].astype('timedelta64[s]')
data.head(20)
```

## Q2: Moving Average

```python
'''
---------------------------------
1. MA næœŸ,æœƒæœ‰n-1å€‹NaN
2. t-(MAæœŸæ•¸-1)~t -> return to t
---------------------------------
index|time_diff_m|time_diff_m_r2|
0    |0          |NaN           |
1    |10         |5             |
2    |20         |15            |
'''

data['time_diff_m_r2']=data.groupby(['shopid'])['time_diff_m'].apply(lambda x : x.rolling(2).mean())
#subset->rolling->return
data.head(20)
```
