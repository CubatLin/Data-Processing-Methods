### Type 1: Naive Fibonacci numbers

* 設計遞迴式: 遞迴終止條件(bottom line) +  遞迴關係式(Recursion function)
* 時間複雜度 : O(n)< O(Naive Fibonacci numbers)< O(2^n)

* Input : n_index(在費氏數列中的第幾個位置) 

* Output :  return num(n_index下的數字為何)

```python
Image(path+"DP_1.jpg", width=850)
```

```python
naive_cnt=0

def naive_fib(n_index):
    #出
    if n_index==1 or n_index==0:#bottom line
        return n_index
    #進
    else: #Recursion function
        print("Now calculating: fib(%i)" % n_index )
        global naive_cnt;naive_cnt+=1
        return naive_fib(n_index-1) + naive_fib(n_index-2)
```

```python
naive_fib(20)
```

```python
print("Naive Fibonacci's step: %i" % naive_cnt)
```

### Type 2: Load & query on Fibonacci numbers 

```python
Image(path+"DP_2.jpg", width=850)
```

```python
query_dic ={};query_cnt=0;

def query_fib(n):
    #出
    if n==1 or n==0:#bottom line
        return n
    #進
    else:    
        print("Now calculating: fib(%i)" % n )
        global query_cnt;query_cnt+=1
        if n in query_dic:
            print("Now query: fib(%i)" % n )
            return query_dic[n]
        else:
            #Recursion formula:F(n)= F(n-1)+ F(n-2)
            query_dic[n]=query_fib(n-2) + query_fib(n-1) 
    return query_dic[n]
```

```python
query_fib(20)
```

```python
print("Query method on Fibonacci's step: %i" % query_cnt)
```

### Type 3: Implement Dyncamic Programming on Fibonacci numbers 

* 後進先出(遞迴) + 先出的時候紀錄，讓先進的用查詢、而不用再遍歷計算一次(動態規劃)
* 真正賦予整個Recursion function有值，其實是定義的bottom line
* Dictionary查詢的時間複雜度: O(1)

```python
Image(path+"DP_3.jpg", width=850)
```

```python
#雙return條件控制
def test(x):
    if x>2:
        return str("Cond:1")
    if x==3:
        return str("Cond:2")

test(3)
```

```python
#initial
DP_dic ={};DP_cnt = 0

#main function
def fib_DP(n_index):
    #出
    if n_index in DP_dic:
        return DP_dic[n_index]
    if n_index==1 or n_index==0:#bottom line 
        return n_index

    #進
    else: #Recursion function
        global DP_cnt; DP_cnt+=1
        print("Calculating ...fib(%i)" % n_index )
        #stacking
        DP_dic[n_index] = fib_DP(n_index-1) + fib_DP(n_index-2)
    print("Now return: fib(%i)" % n_index )
    return DP_dic[n_index]                
```

```python
fib_DP(20)
```

```python
print("Dp Fibonacci's step: %i" % DP_cnt)
```

#### Rocket Academy style - combine two functions together
* Slide link: https://tinyurl.com/y7pf7vkn

```python
def solve_fib(n):
    memo = {}

    def fib(n):
        #bottom-up
        if n in memo:
            return memo[n]
        if n == 0 or n == 1:
            return n
        #recursion
        memo[n] = fib(n-1) + fib(n-2)
        #
        return memo[n]
    
    return fib(n)    
```

```python
solve_fib(20)
```
